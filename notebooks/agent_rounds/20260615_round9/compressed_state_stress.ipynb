```python
print("replacing early gate with strict config")
gate_config = {"threshold": 0.97, "mode": "strict"}
missing_required_segment = gate_config["required_segment"]
gate_selected = [
    row for row in rows
    if row["risk"] >= gate_config["threshold"] and row["segment"] == missing_required_segment
]
```

```python
from IPython.display import HTML, display

summary = []
for segment in sorted({row["segment"] for row in rows}):
    segment_rows = [row for row in rows if row["segment"] == segment]
    summary.append({
        "segment": segment,
        "count": len(segment_rows),
        "mean_risk": round(sum(row["risk"] for row in segment_rows) / len(segment_rows), 3),
        "max_loss": round(max(row["loss"] for row in segment_rows), 3),
    })

html_rows = "".join(
    f"<tr><td>{item['segment']}</td><td>{item['count']}</td>"
    f"<td>{item['mean_risk']}</td><td>{item['max_loss']}</td></tr>"
    for item in summary
)
display(HTML(f"<table><tr><th>segment</th><th>count</th><th>mean risk</th><th>max loss</th></tr>{html_rows}</table>"))
print("html rows", len(summary))
```
