```python
from pathlib import Path
import json
import os

artifact_dir = Path(os.environ["NL_ARTIFACT_DIR"])
artifact_dir.mkdir(parents=True, exist_ok=True)
payload = {
    "cwd": str(Path.cwd()),
    "nl_artifact_dir_env": os.environ["NL_ARTIFACT_DIR"],
    "nl_notebook_dir_env": os.environ.get("NL_NOTEBOOK_DIR"),
    "artifact_dir_resolved": str(artifact_dir.resolve()),
}
(artifact_dir / "path_probe.json").write_text(json.dumps(payload, indent=2) + "\n")
print(json.dumps(payload, indent=2))
```
