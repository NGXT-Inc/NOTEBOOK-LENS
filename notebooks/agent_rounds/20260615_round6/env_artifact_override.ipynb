```python
import json
import os
from pathlib import Path

root = Path(os.environ["NL_ARTIFACT_DIR"])
inferred = Path("artifacts/agent_rounds/20260615_round6/env_artifact_override")

root.mkdir(parents=True, exist_ok=True)
(root / "nested").mkdir(parents=True, exist_ok=True)
inferred.mkdir(parents=True, exist_ok=True)

(root / "created.txt").write_text("external created by notebook\n", encoding="utf-8")
(root / "modified.txt").write_text("external canonical content\n", encoding="utf-8")
(root / "delete_me.txt").write_text("external delete target\n", encoding="utf-8")
(root / "nested" / "scope.json").write_text(
    json.dumps(
        {
            "scope": "NL_ARTIFACT_DIR",
            "round": "20260615_round6",
            "scenario": "env_artifact_override",
        },
        indent=2,
        sort_keys=True,
    )
    + "\n",
    encoding="utf-8",
)

(inferred / "inferred_only.txt").write_text(
    "inferred path should not be tracked when NL_ARTIFACT_DIR is set\n",
    encoding="utf-8",
)

external_files = sorted(
    path.relative_to(root).as_posix() for path in root.rglob("*") if path.is_file()
)

print(f"env_root={root}")
print(f"inferred_root={inferred}")
print(f"external_files={external_files}")
print(f"inferred_file_exists={(inferred / 'inferred_only.txt').exists()}")
```
