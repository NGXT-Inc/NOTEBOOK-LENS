This notebook was generated through the Notebook Lens CLI to check that:

- markdown cells survive as standard notebook content
- stdout stream output is preserved
- rich display output includes HTML and image data
- the local read-only HTTP UI can serialize the notebook over loopback


```python
print("stdout compatibility probe")
for index, label in enumerate(["markdown", "stdout", "rich-output"], start=1):
    print(f"{index}: {label}")
```

```python
import struct
import zlib

from IPython.display import HTML, Image, display


def png_chunk(kind, data):
    payload = kind + data
    checksum = zlib.crc32(payload) & 0xFFFFFFFF
    return struct.pack(">I", len(data)) + payload + struct.pack(">I", checksum)


width = 8
height = 8
rows = []
for y in range(height):
    row = bytearray([0])
    for x in range(width):
        if (x + y) % 2:
            row.extend((15, 118, 110))
        else:
            row.extend((245, 247, 250))
    rows.append(bytes(row))

png_bytes = (
    b"\x89PNG\r\n\x1a\n"
    + png_chunk(b"IHDR", struct.pack(">IIBBBBB", width, height, 8, 2, 0, 0, 0))
    + png_chunk(b"IDAT", zlib.compress(b"".join(rows)))
    + png_chunk(b"IEND", b"")
)

display(
    HTML(
        "<table><thead><tr><th>format</th><th>status</th></tr></thead>"
        "<tbody><tr><td>text/html</td><td>rendered</td></tr></tbody></table>"
    )
)
display(Image(data=png_bytes, format="png"))
print(f"embedded png bytes: {len(png_bytes)}")
```
