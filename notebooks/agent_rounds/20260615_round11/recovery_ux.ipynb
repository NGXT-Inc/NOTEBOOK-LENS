# Recovery UX probe

This notebook exercises Notebook Lens recovery behavior around explicit resets,
external notebook-file edits, and failed cells. The goal is to inspect the
agent-facing command text and follow the returned recovery guidance without
assuming hidden state.


```python
import os

x = 10
round11_events = ["seed"]

print("seed_x", x)
print("kernel_pid", os.getpid())
print("events", "|".join(round11_events))
```

```python
round11_events.append("after_reset")

print("after_reset_x_plus", x + 1)
print("event_count", len(round11_events))
print("events", "|".join(round11_events))
```

## External notebook edit

This markdown cell was appended by a simulated editor outside Notebook Lens while the CLI-managed kernel was still live.

```python
round11_events.append("after_external")

print("after_external_x", x)
print("event_count", len(round11_events))
print("events", "|".join(round11_events))
```

```python
round11_events.append("failure_repaired")
failure_payload = {"phase": "repaired", "x": x}

print("failure_repaired", failure_payload)
print("events_after_repair", "|".join(round11_events))
```

```python
round11_events.append("after_repair")

print("final_events", "|".join(round11_events))
print("final_x", x)
```
