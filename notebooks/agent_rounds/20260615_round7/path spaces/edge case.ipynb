```python
threshold = 10
value = 12
decision = "ship" if value >= threshold else "hold"

print(f"seed_threshold={threshold}")
print(f"seed_value={value}")
print(f"seed_decision={decision}")
```

The seed value is **12**, which is above the threshold of **10**.

Decision: **ship**. The downstream score recomputes to **14** after `run-clean`.


```python
score = value + 2
downstream_decision = "ship" if score >= threshold else "hold"

print(f"downstream_score={score}")
print(f"downstream_decision={downstream_decision}")
```

```python
print(f"final_value={value}")
print(f"final_score={score}")
print(f"final_decision={downstream_decision}")
```
