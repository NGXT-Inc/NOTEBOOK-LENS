```python
import hashlib
import json
import os
import uuid
from datetime import datetime, timezone
from pathlib import Path

ROOT_REL = Path("artifacts/agent_rounds/20260615_round10/env_sync_edges")
PHASE = os.environ.get("NOTEBOOK_LENS_ENV_SYNC_PHASE", "missing_phase")
ENV_KEYS = [
    "NL_ARTIFACT_DIR",
    "ML_ARTIFACT_DIR",
    "HF_HOME",
    "HF_DATASETS_CACHE",
    "HF_ENDPOINT",
    "HF_TOKEN",
    "HF_API_KEY",
    "ML_SECRET_KEY",
    "NOTEBOOK_LENS_SECRET_TOKEN",
]
SECRET_KEYS = {
    "HF_TOKEN",
    "HF_API_KEY",
    "ML_SECRET_KEY",
    "NOTEBOOK_LENS_SECRET_TOKEN",
}


def digest(value):
    if value is None:
        return None
    return hashlib.sha256(value.encode("utf-8")).hexdigest()[:12]


def sanitize_env_value(key):
    value = os.environ.get(key)
    if value is None:
        return {"present": False}
    if key in SECRET_KEYS:
        return {"present": True, "sha256_12": digest(value), "length": len(value)}
    if key.endswith("_DIR") or key in {"HF_HOME", "HF_DATASETS_CACHE"}:
        return {
            "present": True,
            "basename": Path(value).name,
            "sha256_12": digest(value),
        }
    return {"present": True, "value": value, "sha256_12": digest(value)}


marker = globals().setdefault("_ENV_SYNC_EDGE_MARKER", uuid.uuid4().hex[:12])
history = globals().setdefault("_ENV_SYNC_EDGE_HISTORY", [])
snapshot = {
    "phase": PHASE,
    "pid": os.getpid(),
    "marker": marker,
    "captured_at": datetime.now(timezone.utc).isoformat(),
    "env": {key: sanitize_env_value(key) for key in ENV_KEYS},
}
history.append(snapshot)

artifact_root = Path(os.environ.get("NL_ARTIFACT_DIR") or ROOT_REL)
artifact_root.mkdir(parents=True, exist_ok=True)
(artifact_root / f"{PHASE}_snapshot.json").write_text(
    json.dumps(snapshot, indent=2, sort_keys=True) + "\n",
    encoding="utf-8",
)

if os.environ.get("ML_ARTIFACT_DIR"):
    ml_root = Path(os.environ["ML_ARTIFACT_DIR"])
    ml_root.mkdir(parents=True, exist_ok=True)
    (ml_root / f"{PHASE}_ml_seen.json").write_text(
        json.dumps(
            {"phase": PHASE, "pid": snapshot["pid"], "marker": marker},
            indent=2,
            sort_keys=True,
        )
        + "\n",
        encoding="utf-8",
    )

print(
    json.dumps(
        {
            "phase": PHASE,
            "pid": snapshot["pid"],
            "marker": marker,
            "history_count": len(history),
            "env": snapshot["env"],
        },
        indent=2,
        sort_keys=True,
    )
)
```

```python
import json
import os
from pathlib import Path

ROOT_REL = Path("artifacts/agent_rounds/20260615_round10/env_sync_edges")
EDGE_ASSERT = os.environ.get("NOTEBOOK_LENS_EDGE_ASSERT") == "1"
EXPECTED_PHASES = ["phase_a", "phase_b", "phase_removed"]
CHANGE_KEYS = [
    "NL_ARTIFACT_DIR",
    "ML_ARTIFACT_DIR",
    "HF_HOME",
    "HF_DATASETS_CACHE",
    "HF_ENDPOINT",
    "HF_TOKEN",
    "HF_API_KEY",
    "ML_SECRET_KEY",
    "NOTEBOOK_LENS_SECRET_TOKEN",
]

history = list(globals().get("_ENV_SYNC_EDGE_HISTORY", []))
summary = {
    "mode": "assert" if EDGE_ASSERT else "clean_rerun_redaction_only",
    "history_count": len(history),
    "phases": [item.get("phase") for item in history],
    "pid_count": len({item.get("pid") for item in history}),
    "marker_count": len({item.get("marker") for item in history}),
}


def env_entry(entry, key):
    return entry["env"][key]


def fingerprint(entry, key):
    current = env_entry(entry, key)
    return current.get("sha256_12") or current.get("value") or current.get("basename")


if EDGE_ASSERT:
    entries = {}
    for phase in EXPECTED_PHASES:
        matches = [item for item in history if item.get("phase") == phase]
        if not matches:
            raise AssertionError(f"missing phase in live-kernel history: {phase}")
        entries[phase] = matches[-1]

    checks = {
        "same_kernel_pid": len({entries[phase]["pid"] for phase in EXPECTED_PHASES}) == 1,
        "same_kernel_marker": len({entries[phase]["marker"] for phase in EXPECTED_PHASES}) == 1,
    }
    for key in CHANGE_KEYS:
        checks[f"{key}_phase_a_present"] = env_entry(entries["phase_a"], key)["present"]
        checks[f"{key}_phase_b_present"] = env_entry(entries["phase_b"], key)["present"]
        checks[f"{key}_changed_a_to_b"] = fingerprint(entries["phase_a"], key) != fingerprint(
            entries["phase_b"], key
        )
        checks[f"{key}_removed"] = not env_entry(entries["phase_removed"], key)["present"]

    failures = [name for name, ok in checks.items() if not ok]
    verdict = {
        "status": "pass" if not failures else "fail",
        "failures": failures,
        "summary": summary,
        "checks": checks,
    }
    ROOT_REL.mkdir(parents=True, exist_ok=True)
    (ROOT_REL / "incremental_env_sync_verdict.json").write_text(
        json.dumps(verdict, indent=2, sort_keys=True) + "\n",
        encoding="utf-8",
    )
    if failures:
        raise AssertionError(json.dumps(verdict, indent=2, sort_keys=True))
    print(json.dumps(verdict, indent=2, sort_keys=True))
else:
    print(json.dumps({"status": "skipped", "summary": summary}, indent=2, sort_keys=True))
```
