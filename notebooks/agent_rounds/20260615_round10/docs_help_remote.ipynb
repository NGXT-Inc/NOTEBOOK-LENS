# Findings summary

`--file <path>` is the safest documented remote workflow because it does not
depend on SSH stdin forwarding. `--file -` works when stdin is actually piped,
but an empty stdin stream is rejected before a notebook cell is created.

Notebook path resolution also matched the README guidance in this run:
experiment-root paths, paths relative to `NL_NOTEBOOK_DIR`, and absolute paths
inside the notebook directory all resolved to the same notebook.
