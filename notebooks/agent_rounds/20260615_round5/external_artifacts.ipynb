```python
from pathlib import Path
import json
import os

root = Path(os.environ["NL_ARTIFACT_DIR"])
root.mkdir(parents=True, exist_ok=True)

(root / "created_by_run_cell.txt").write_text("created by run-cell\n", encoding="utf-8")
(root / "modified_by_run_cell.txt").write_text("initial content\n", encoding="utf-8")
(root / "deleted_by_run_cell.txt").write_text("will be deleted\n", encoding="utf-8")

nested = root / "nested"
nested.mkdir(exist_ok=True)
(nested / "scope.json").write_text(
    json.dumps(
        {
            "root": str(root),
            "secret_env_present": "NL_FAKE_SECRET" in os.environ,
        },
        sort_keys=True,
    )
    + "\n",
    encoding="utf-8",
)

print("artifact_root", root)
print("secret_env_present", "NL_FAKE_SECRET" in os.environ)
```

```python
from pathlib import Path
import os

root = Path(os.environ["NL_ARTIFACT_DIR"])

(root / "modified_by_run_cell.txt").write_text(
    f"modified by run-cell\nroot={root}\n",
    encoding="utf-8",
)

delete_target = root / "deleted_by_run_cell.txt"
deleted_existed = delete_target.exists()
if deleted_existed:
    delete_target.unlink()

(root / "nested" / "second_cell.txt").write_text("second cell artifact\n", encoding="utf-8")

counter_path = root / "clean_counter.txt"
try:
    counter = int(counter_path.read_text(encoding="utf-8").strip())
except (FileNotFoundError, ValueError):
    counter = 0
counter_path.write_text(f"{counter + 1}\n", encoding="utf-8")

print("modified", root / "modified_by_run_cell.txt")
print("deleted_by_run_cell_existed", deleted_existed)
print("clean_counter", counter + 1)
```
