```python
from pathlib import Path
import json

scope = Path("artifacts/round5_general/generic_scope")
noise = Path("artifacts/round5_general/other_noise")
scope.mkdir(parents=True, exist_ok=True)
noise.mkdir(parents=True, exist_ok=True)

metrics = {
    "round": 5,
    "scenario": "generic non-agent-round inferred scope",
    "notebook": "notebooks/round5_general/generic_scope.ipynb",
    "expected_scope": "artifacts/round5_general/generic_scope",
}

(scope / "metrics.json").write_text(json.dumps(metrics, indent=2) + "\n")
(scope / "summary.txt").write_text(
    "intended artifact for generic non-agent-round inferred scope\n"
)

(noise / "noise.txt").write_text("deliberate sibling noise from run-cell\n")
(noise / "noise.json").write_text(json.dumps({"noise": True}, indent=2) + "\n")

print("intended artifacts:")
for path in sorted(scope.iterdir()):
    print(f"- {path}")

print("noise artifacts:")
for path in sorted(noise.iterdir()):
    print(f"- {path}")
```
